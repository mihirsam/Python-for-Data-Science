```python
file = open("98-0.txt","r")
stopword = open("stopwords.txt", "r")
```

```python
word_cloud = {}
```

```python
for word in file:
    words = []
    words = word.split(' ')
    
    for i in range(0, len(words)):
        if words[i] not in word_cloud.keys():
            word_cloud[words[i]] = 1
        else:
            word_cloud[words[i]] += 1
        
```
